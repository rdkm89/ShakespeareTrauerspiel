```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
data = pd.read_csv("ratio_all_neg_emo.csv", sep=";")
```

```python
data = data.sort_values(by = "ratio", ascending=False)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(18.5, 10.5)

colors = {'S':'mediumseagreen', 'N':'orange'}

ax.bar(data['filenames'], data['ratio'], color=data['author'].apply(lambda x: colors[x]))

#ax.bar(shakes["filenames"], shakes['ratio'], color="mediumseagreen", label='Shakespeare')
#ax.bar(non_shakes["filenames"], non_shakes['ratio'], color="orange", label='Non-Shakespeare')

shakes = mpatches.Patch(color='mediumseagreen', label='Shakespeare')
non_shakes = mpatches.Patch(color='orange', label='Non-Shakespeare')
plt.legend(handles=[shakes, non_shakes])

plt.xticks(rotation='vertical')
plt.tick_params(bottom=False, left=False)

#ax.invert_yaxis() 
ax.set_title('All Negative Emotions')
plt.box(False)

plt.savefig("all_neg_emo_distribution.png")

plt.show()
```
